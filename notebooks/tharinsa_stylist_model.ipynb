Day 3

```python
prompts = {
    "casual_outings": {
        "beach_trip": "Suggest 3 casual outfits for a beach trip",
        "park_day": "Suggest 2 casual outfits for a day at the park"
    },
    "color_matching": {
        "olive_green_pants": "What colors go well with olive green pants?",
        "navy_blazer": "What colors go well with a navy blazer?"
    }
}
```

```python
generator = pipeline("text-generation", model="gpt2")

for category, sub_prompts in prompts.items():
    for scenario, prompt in sub_prompts.items():
        output = generator(prompt, max_new_tokens=50, truncation=True)
        print(f"{category} - {scenario}:")
        print(output[0]['generated_text'])
        print()
```

```python
from transformers import pipeline
import re

generator = pipeline("text-generation", model="gpt2")

# This function will format the output
def format_ai_output(response):
   
    cleaned = re.sub(r'[^\w\s.,!?]', '', response) # Cleaning unwanted characters 
    sentences = cleaned.split('.') # Trimming incomplete sentences
    complete_sentences = [s for s in sentences if s.strip() and s.strip()[-1] in ['.', '!', '?']]
    if complete_sentences:
        return '.'.join(complete_sentences) + '.'
    else:
        return cleaned

prompt = "What should i wear for a night out in Paris: "
output = generator(prompt, max_new_tokens=50, truncation=True)

formatted_output = format_ai_output(output[0]['generated_text'])
print("\nOutput:", formatted_output)
```
